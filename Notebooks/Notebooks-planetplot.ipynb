```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
```

```python
df = pd.read_csv("../work/HSRA Research Sheets - Planets.csv")
df.rename(columns = {'pl_radius (Earth radius)':'radius', 'pl_mass (Earth mass)':'mass'}, inplace = True)

df['radius'] = 6371000 * df.radius

df['mass'] = 5.9722e24 * df.mass

df['volume'] = (4/3) * math.pi * ((df.radius)**3)

df['density'] = df.mass / df.volume

df['radius'] = np.log10(df.radius)

df['density'] = np.log10(df.density)

#The following code is the plotting code and the lin reg determining code

mpl.scatter(df['radius'], df['density'], s=5)
mpl.xlabel('Log (Radius in $m$)')
mpl.ylabel('Log (Density in $kg/m^3$)')
z = np.polyfit(df['radius'], df['density'], 1)
p = np.poly1d(z)
yp = p(df['radius'])
ym = np.mean(df['density'])
sst = np.sum((df['density'] - ym)**2)
ssr = np.sum((df['density'] - yp)**2)
r2 = round(1 - (ssr/sst), 3)
mpl.plot(df['radius'], p(df['radius']), color = 'black', label = 'Lin Reg Trend')
mpl.show()
print("R^2:" + str(r2))
```
